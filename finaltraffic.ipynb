IMPORTING LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

READING IMAGE

```python
im_cv = cv2.imread("public_test_images\maze_1.png")
plt.imshow(im_cv)
plt.show()
```

```python
img = cv2.cvtColor(im_cv, cv2.COLOR_BGR2RGB)
array=np.array(img)
plt.imshow(img)
plt.show
```

FUNCTION TO KNOW NODE

```python
def node(i,j):
 if(i==100):
    x='A'
 if(i==200):
    x='B'
 if(i==300):
    x='C'
 if(i==400):
    x='D'
 if(i==500):
    x='E'
 if(i==600):
    x='F'
 if(i==700):
    x='G'
 Y=(x+str(int(j/100))) 
 return Y
```

PRINTING PLOT VALUE OF TRAFFIC LIGHT

```python
red=[255,0,0]
i=100
j=100
points = []
for i in {100,200,300,400,500,600,700}:
    for j in {100,200,300,400,500,600,700}:
             abcd=array[i,j]
             equal_arrays = (red == abcd).all()
             if equal_arrays==1:
                print(j,i)
                points.append((j, i))
```

```python
mask = np.zeros_like(img)

for point in points:
    cv2.circle(mask, tuple(point), 5, (255, 255, 255), -1)

masked_img = cv2.bitwise_and(img, mask)

plt.imshow(masked_img)
plt.show()
```

FINAL VALUE OF VERTICAL SPACE

```python
white=[255,255,255]
i=100
j=100
ranges = []
for i in {100,200,300,400,500,600,700}:
        for j in {110,210,310,410,510,610}:
             abcd=array[j,i]
             equal_arrays = (white == abcd).all()
             if equal_arrays==1:
                y= node(i,j)
                z=node(i,j+100)
                print(y+"-"+z)
                ranges.append(((i, j-10), (i, j+90)))
```

```python
mask = np.zeros_like(img)

for start, end in ranges:
    x_start, y_start = start
    x_end, y_end = end
    cv2.rectangle(mask, (x_start, y_start), (x_end, y_end), (255, 255, 255), -1)

masked_img = cv2.bitwise_and(img, mask)

plt.imshow(masked_img)
plt.show()
```
